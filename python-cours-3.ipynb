```python
# 1er exercice
def conjuge(verbe):
    pronom = {"je / j'": "e", "tu": "es", "il": "e", "nous": "ons", "vous": "ez", "ils": "ent"}
    if verbe.count('g') > 0:
        pronom["nous"] = "eons"
    for cle, valeur in pronom.items():
        print(cle + " " + verbe[:-2] + valeur)


verbe = input("veuillez saisir un verbe du 1er groupe : ")
conjuge(verbe)
```

```python
# 2eme exercice
def distance_de_hamming(mot1, mot2):
    cpt = 0
    for i in mot1:
        if mot2.count(i) > 0:
            cpt += 1
    return len(mot1) - cpt


meme_longueur = True
while meme_longueur:
    mot1 = input("veuillez saisir le premier mot : ")
    mot2 = input("veuillez saisir le deuxième mot : ")
    if len(mot1) != len(mot2):
        print("ERREUR : Les mots saisies ne sont pas de même longueur ! Recommencez")
    else:
        meme_longueur = False

print(f"La distance de Hamming est de : {distance_de_hamming(mot1, mot2)}")
```

```python

# exercice 6
def compteur_consonne(mot):
    cpt = 0
    consonnes = ["B", "C", "D", "F", "G", "H", "J", "K", "L", "M", "N", "P", "Q", "R", "S", "T", "V", "W", "X", "Z"]
    for i in range(0, len(mot)):
        for j in range(0, len(consonnes)):
            if mot.upper()[i] == consonnes[j]:
                cpt += 1
    return cpt


mot = input("Veuillez saisir un mot : ")
print(f"Le nombre de consonne(s) dans {mot} est de : {compteur_consonne(mot)}")
```

```python
# exercice 7

def encode(nom):
    nom = nom.upper()
    chaine = ""
    print("Le nom codé est le suivant :")
    for i in range(0, len(nom)):
        if chr(ord(nom[i])) == 'X':
            chaine += 'a'
        elif chr(ord(nom[i])) == 'Y':
            chaine += 'b'
        elif chr(ord(nom[i])) == 'Z':
            chaine += 'c'
        else:    
            chaine += chr(ord(nom[i]) + 3)
    return chaine

def decode(nom):
    nom = nom.upper()
    chaine = ""
    print("Le nom decocodé est le suivant :")
    for i in range(0, len(nom)):
        if chr(ord(nom[i])) == 'A':
            chaine += 'x'
        elif chr(ord(nom[i])) == 'B':
            chaine += 'y'
        elif chr(ord(nom[i])) == 'C':
            chaine += 'z'
        else:
            chaine += chr(ord(nom[i]) - 3)
    return chaine
        
name1 = input("Veuillez entrez votre nom a encoder : ")
print(encode(name1))

name2 = input("Veuillez entrez votre nom a decoder : ")
print(decode(name2))
```
